# chapter_10_sequence_model
* **不同的RNN应用类型**
  * one-to-one: data transform
  * one-to-many: music generation
  * many-to-one: text classification
  * many-to-many

image.png

```python
from torch import nn
import torch
```

```python
model_rnn = nn.RNN(
    input_size=4,       #input data channel
    hidden_size=128,    #rnn hidden dim 64-1024
    bias=False,
    num_layers=1,       #rnn 层数
    bidirectional=False,
)
```

```python
# [L, N, H_{in}]
# L:sequence length
# N:batch size
# H_{in}:input channel

x = torch.randn(size=(31, 8, 4))
output, hn = model_rnn(x)

print(output.size()) #每一步的时间序列输出
print(hn.size())     #最后一步的时间序列的输出
```

* **backpropagation through time(延时梯度反向传播)**
  * 过长的时间步导致梯度消失
  * 不当的初始化权重导致梯度爆炸
